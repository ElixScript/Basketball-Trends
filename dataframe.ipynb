# seaborn

```python
# intro to seaborn 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4
import warnings
warnings.filterwarnings('ignore')
```

```python
# distribution :
vis1 = sns.displot(stats['InternetUsers'])
```

```python
vis1 = sns.displot(stats['InternetUsers'], bins = 30)
```

```python
# boxplot 
vis2 = sns.boxplot(data=stats,x = 'IncomeGroup',y= 'BirthRate')
```

```python
# visualising with seaborn 
vis3 = sns.lmplot( data = stats, x = 'InternetUsers',y='BirthRate',fit_reg = False)
```

```python
vis3 = sns.lmplot( data = stats, x = 'InternetUsers',y='BirthRate', 
                  fit_reg = False, hue = 'IncomeGroup')
```

```python
vis3 = sns.lmplot( data = stats, x = 'InternetUsers',y='BirthRate', 
                  fit_reg = False, hue = 'IncomeGroup', scatter_kws = {'s' : 100})
```
